```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section
#we can see that 21 columns in the dataset, We can drop them using drop function.
task.drop(['budget_adj','revenue_adj','overview','imdb_id','homepage','tagline'],axis =1,inplace = True)
print("Afetr Removing Unused Columns (Rows,Columns) : ",task.shape)

```

```python

# Checking for an zero values in the budget and revenue columns
print("Rows With Zero Values In The Budget Column:",task[(task['budget']==0)].shape[0])
print("Rows With Zero Values In The Revenue Column:",task[(task['revenue']==0)].shape[0])
```

```python
#map all the rows of genres in a list.
genres_details = list(map(str,(task['genres'])))
# print (genres_details)
genres = ['Adventure', 'Science Fiction', 'Fantasy', 'Crime', 'Western', 'Family','nan', 'Animation','War','Mystery','Romance','TV Movie','Action', 'Thriller','Comedy','Drama' , 'History', 'Music', 'Horror', 'Documentary', 'Foreign']

#make the numpy array of year and popularity which contain all the rows of release_year and popularity column. 
year = np.array(task['release_year'])
popularity = np.array(task['popularity'])

#make a null dataframe which indexs are genres and columns are years.
popularity_df = pd.DataFrame(index = genres, columns = range(1960, 2016))
#change all the values of the dataframe from NAN to zero.
popularity_df = popularity_df.fillna(value = 0.0)

z = 0
for i in genres_details:
    split_genre = list(map(str,i.split('|')))
#     print(split_genre)
    popularity_df.loc[split_genre, year[z]] = popularity_df.loc[split_genre, year[z]] + popularity[z]
    z+=1
```

```python
#calculate the standard deviation for the accurate results.
def calculate_std(x):
    return (x-x.mean())/x.std(ddof=0)

popular_genre = calculate_std(popularity_df)
popular_genre.head(10)
```

```python
#plot the barh plot of the standardised data.
popular_genre.iloc[10:,53:].plot(kind='bar',figsize = (15,6),fontsize=13)

#setup the title and labels of the plot.
plt.title("Most Popular Genre Over Year To Year",fontsize=12)
plt.xlabel("Popularity (Standerd Units)",fontsize=14)
plt.ylabel("Genres",fontsize = 14)
sns.set_style("whitegrid")
```

```python
#How the popularity of the genre differ year by year.
sns.set_style("whitegrid")
#make a subplot of size 3,3.
fig, ax = plt.subplots(3,3,figsize = (16,10))
#set the title of the subplot.
fig.suptitle('Genres Popularity Over Year To Year',fontsize = 16)


#accordig to the previous plot the 'Drama','Action' ,'Comedy','Horror','Science Fiction', 'Romance', 'Music','Adventure' and 'Crime'.
popular_genre.loc['Drama'].plot(label = "Drama",color = 'red',ax = ax[0][0],legend=True)
popular_genre.loc['Action'].plot(label = "Action",color='g',ax = ax[0][1],legend=True)
popular_genre.loc['Comedy'].plot(label = "Comedy",color='#00788a',ax = ax[0][2],legend=True)
popular_genre.loc['Horror'].plot(label = "Horror",color='#00818a',ax = ax[1][0],legend=True)
popular_genre.loc['Science Fiction'].plot(label = "Science Fiction",color='#6f6600',ax = ax[1][1],legend=True)
popular_genre.loc['Romance'].plot(label = "Romance",color='#1a2c5b',ax = ax[1][2],legend=True)
popular_genre.loc['Music'].plot(label = "Music",color='#db3b61',ax = ax[2][0],legend=True)
popular_genre.loc['Adventure'].plot(label = "Adventure",color='#08c299',ax = ax[2][1],legend=True)
popular_genre.loc['Crime'].plot(label = "Crime",color='c',ax = ax[2][2],legend=True)
```

```python
def plot_correlation_map( df ):
    corr = df.corr()
    fig , ax = plt.subplots(figsize =( 12 , 10 ) )
    fig.suptitle('map views',fontsize = 16)
    cmap = sns.diverging_palette( 240 , 10 , as_cmap = True )
    fig = sns.heatmap(corr,cmap = cmap,square=True, cbar_kws={ 'shrink' : .9 }, ax=ax, annot = True, annot_kws = { 'fontsize' : 12 })
    
```

```python
plot_correlation_map(task[['popularity','budget','revenue','runtime','vote_average']])
```
